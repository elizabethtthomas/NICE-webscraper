### NICE Guideline Webscraper

```python
# Install beautiful soup library 

import os
from bs4 import BeautifulSoup as bs
import requests

# import requests
from urllib import request
from urllib.parse import urljoin
from urllib.parse import urlparse
```

```python
# list the relevant links 
conditions = ["https://www.nice.org.uk/guidance/cg57/evidence", "https://www.nice.org.uk/guidance/ng143/evidence", "https://www.nice.org.uk/guidance/CG116/evidence", "https://www.nice.org.uk/guidance/ng1/evidence"]
    
for condition in conditions: 
#  html = requests.get(condition)
    soup = bs(requests.get(condition).content, "html.parser")
  
 #  get the links 
    for link in soup.find_all('a'):
        current_link = link.get('href')
        print(current_link)
        split_link = urlparse(current_link).path.rstrip('/').split('/')[-1]
        
 # identify links with pdfs, filename as final part of URL, download as pdf        
        if 'pdf' in split_link:
            filename = split_link + ".pdf"
            with open(filename, 'wb') as f:
                full_link = "https://www.nice.org.uk" + current_link
                print(full_link)
                file = requests.get(full_link)
                f.write(file.content)
```
